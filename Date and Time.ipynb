```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("Date and Time").getOrCreate()
```

```python
df = spark.read.csv('Python-and-Spark-for-Big-Data-master/Spark_DataFrames/appl_stock.csv', 
                   header=True, inferSchema=True)
```

```python
df.head(1)
```

```python
df.show(5)
```

```python
from pyspark.sql.functions import (hour, month, year, format_number,
                                   dayofmonth, dayofweek, dayofyear,
                                  weekofyear, date_format)
```

```python
df.select(dayofmonth(df['Date'])).show(5)
```

```python
#df.select(year(df['Date'])).show()
df.withColumn("Year", year(df['Date'])).show(5)
```

```python
newdf = df.withColumn("Year", year(df['Date']))
```

```python
newdf.groupBy('Year').mean().show()
```

```python
newdf.select(['Year', 'Close']).groupBy('Year').mean().show()
```

```python
result = newdf.groupBy('Year').mean().select(['Year','avg(Close)' ])
result.show()
```

```python
result.select(['Year', format_number('avg(Close)', 2).alias("Avg Close")]).show()
```
